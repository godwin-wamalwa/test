```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/onaio/ona-tech/master/data/water_points.json")
pd.options.display.max_rows =100
pd.set_option('display.max_columns', 200)
df.columns
df.T.head(20)
cv = df.filter(regex='communities_villages')
wf = df.filter(regex='water_functioning')
x=pd.DataFrame(cv)
y=pd.DataFrame(wf)
```

```python
cv=pd.DataFrame(x.columns.str.split(':').str[1].tolist(),columns=['communities_villages'])
cv['communities_villages'] = cv['communities_villages'].str.replace('\d+', '').str.replace('"', "").str.replace('.', "")
countCv = cv['communities_villages'].value_counts()
countCv
```

```python
wf=pd.DataFrame(y.columns.str.split(':').str[1].tolist(),columns=['water_functioning'])
wf['water_functioning'] = wf['water_functioning'].str.replace('\d+', '').str.replace('"', "").str.replace('.', "")
countWf = wf['water_functioning'].value_counts()
countWf
```

```python
frames = [cv, wf]
  
result = pd.concat(frames)
```

```python
result.apply(pd.Series.value_counts)
```
